**Exploratory Data Analysis**

Gender

```python
ax = sns.countplot(data=df,x = "Gender")

for bars in ax.containers:
  ax.bar_label(bars)
```

```python
df.groupby(["Gender"],as_index=False)["Amount"].sum().sort_values(by="Amount",ascending=False)
```

```python
sales_gen = df.groupby(["Gender"],as_index=False)["Amount"].sum().sort_values(by="Amount",ascending=False)

sns.barplot(data=sales_gen,x = "Gender",y = "Amount")
```

In above graph we can see that most of the buyer are females.

Age

```python
ax = sns.countplot(data=df,x = "Age Group",hue= "Gender")

for bars in ax.containers:
  ax.bar_label(bars)
```

```python
df.groupby(["Age Group"],as_index=False)["Amount"].sum().sort_values(by="Amount",ascending=False)
```

```python
Sales_age = df.groupby(["Age Group"],as_index=False)["Amount"].sum().sort_values(by="Amount",ascending=False)

sns.barplot(data=Sales_age,x= "Age Group",y = "Amount")
```

From above graph we can see that most of the buyers are age group between 26-35 years female.

Status


```python
df.groupby(["State"],as_index=False)["Orders"].sum().sort_values(by="Orders",ascending=False).head(10)
```

```python
#Total orders / sales from top 10 states.
Sales_states = df.groupby(["State"],as_index=False)["Orders"].sum().sort_values(by="Orders",ascending=False).head(10)

sns.set(rc={"figure.figsize" : (20,5)})
sns.barplot(data=Sales_states,x = "State",y = "Orders")
```

```python
#Total amount / Sales from top 10 states.
Sales_states = df.groupby(["State"],as_index=False)["Amount"].sum().sort_values(by="Amount",ascending=False).head(10)

sns.set(rc={"figure.figsize" : (20,5)})
sns.barplot(data=Sales_states,x = "State",y = "Amount")
```

From above graph we can see that most of the orders from Uttar Pradesh,Maharashtra and Karnataka.

Marital_Status

```python
ax = sns.countplot(data=df,x = "Marital_Status")

sns.set(rc = {"figure.figsize":(6,5)})
for bars in ax.containers:
  ax.bar_label(bars)
```

```python
df.groupby(["Marital_Status","Gender"],as_index=False)["Amount"].sum().sort_values(by = "Amount",ascending=False)
```

```python
Sales_status = df.groupby(["Marital_Status","Gender"],as_index=False)["Amount"].sum().sort_values(by = "Amount",ascending=False)

sns.set(rc = {"figure.figsize":(6,5)})
sns.barplot(data=Sales_status,x = "Marital_Status",y = "Amount",hue="Gender")
```

From above graph we can see that most of the buyers are married(Female).

Occupation

```python
ax = sns.countplot(data=df,x = "Occupation")
sns.set(rc = {"figure.figsize":(20,10)})

for bars in ax.containers:
  ax.bar_label(bars)
```

```python
df.groupby(["Occupation"],as_index=False)["Amount"].sum().sort_values(by ="Amount",ascending=False)
```

```python
Sales_states = df.groupby(["Occupation"],as_index=False)["Amount"].sum().sort_values(by ="Amount",ascending=False)

sns.barplot(data=Sales_states,x = "Occupation",y = "Amount")
```

From above graph we see that most of the buyers are working in IT sector,Healthcare,Aviation.

Product category

```python
ax = sns.countplot(data=df,x = "Product_Category")


for bars in ax.containers:
  ax.bar_label(bars)
```

```python
Sales_states = df.groupby(["Product_Category"],as_index=False)["Amount"].sum().sort_values(by ="Amount",ascending=False).head(10)


sns.barplot(data=Sales_states,x = "Product_Category",y = "Amount")
```

From above graph we can see that most of the sold product are from Food,Clothing and Apparel , Electronics and Gadgets.

```python
df.groupby(["Product_ID"],as_index=False)["Orders"].sum().sort_values(by = "Orders",ascending=False)
```

```python
Sales_states = df.groupby(["Product_ID"],as_index=False)["Orders"].sum().sort_values(by = "Orders",ascending=False).head(10)

sns.set(rc = {"figure.figsize": (20,5)})
sns.barplot(data=Sales_states,x = "Product_ID", y= "Orders")
```

##Conclusion:
**Married women age group 26-35 years from UP, Maharashtra and Karnataka working in IT , Healthcare and Aviation are more likely buy product from Food,Clothing and Electronic category.**
